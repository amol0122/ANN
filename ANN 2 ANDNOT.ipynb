```python
#ANDNOT
threshold=1
def andnot_func(dot,threshold):
    for i in range(0,4):
        if dot[i]>=threshold:
            print('1')
        else:
            print('0')
        
andnot_func(dot,threshold)
```

```python
#AND
wt1= np.array([1,1])
print(wt1)
dot1= arr@wt1
print('dot of and :',dot1)
```

```python
#AND
threshold=2
def and_func(dot1,threshold):
    for i in range(0,4):
        if dot1[i]>=threshold:
            print('1')
        else:
            print('0')
        
and_func(dot1,threshold)
```

```python
#OR
wt2=np.array([1,1])
dot2=arr@wt2
print(dot2)
```

```python
#OR
threshold =1
def or_func(dot2,threshold):
    for i in range(0,4):
        if dot2[i]>=threshold:
            print('1')
        else:
            print('0')

or_func(dot2,threshold)          
```
